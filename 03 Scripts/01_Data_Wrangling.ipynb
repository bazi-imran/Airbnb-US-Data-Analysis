# 4. Data Consistency Checks

```python
# Checking for Mixed-type columns
for col in df.columns.tolist():
  weird = (df[[col]].map(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
# Handing mixed data types. 
# Covert name column mix type into string.
df['name'] = df['name'].astype('str')
```

```python
# Covert host_name mixed type into string
df['host_name'] = df['host_name'].astype('str')
```

```python
# Covert host_name mixed type into string
df['neighbourhood_group'] = df['neighbourhood_group'].astype('str')
```

```python
# Covert host_name mixed type into string
df['last_review'] = df['last_review'].astype('str')
```

```python
# Adderssing missing values
# Replace NaN values with O for reviews_per_month column
df['reviews_per_month'].fillna(0, inplace=True)
```

```python
# Replace alternative representations with NaN
df['neighbourhood_group'].replace({'nan': np.nan}, inplace=True)
```

```python
# Replave NaN with Unknown
df['neighbourhood_group'].fillna('unknown', inplace=True)
```

```python
# Replace nan values with Unknown for last_review column
# Check for alternative representations of missing values
# Replace alternative representations with NaN
df['last_review'].replace({'nan': np.nan}, inplace=True)

# Replave NaN with Unknown
df['last_review'].fillna(0, inplace=True)
```
